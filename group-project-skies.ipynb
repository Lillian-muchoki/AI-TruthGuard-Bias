** Essential columns**
1. Bias Detection & AI Fairness

User_Rating → Measures user satisfaction, which may indicate bias trends.

Region → Helps analyze AI fairness across different locations.

Language → Identifies potential bias in multilingual responses.

Device_Type → Checks if AI performs differently across devices.

2. AI Accuracy & Hallucination Detection

Response_Accuracy → Direct metric for AI correctness.

Correction_Needed → Shows cases where AI responses required corrections.

Response_Tokens → Longer responses may have more hallucinations.

3. AI Performance & User Engagement

Session_Duration_sec → Measures how long users interact with AI.

User_Experience_Score → Assesses overall AI interaction quality.
    
Customer_Support_Interactions → Indicates user complaints or confusion.

4. Comparative Analysis (DeepSeek vs. ChatGPT)

AI_Platform → Differentiates between ChatGPT and DeepSeek.

AI_Model_Version → Helps track model improvements over time.
