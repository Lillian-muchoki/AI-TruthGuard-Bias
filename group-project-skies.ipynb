```python
import matplotlib.pyplot as plt
import seaborn as sns

#Plot box plot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 5, i)
    sns.boxplot(y=df_rel[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Define a function to remove outliers using the IQR method
def detect_outliers(df_rel,columns):
   # df_rel=df_copy()
    for col in columns:
        
        
        Q1 = df_rel[columns].quantile(0.25)
        Q3 = df_rel[columns].quantile(0.75)
        IQR = Q3 - Q1

    # Define lower and upper bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers
        df_rel_filtered = df_rel[~((df_rel[columns] < lower_bound) | (df_rel[columns] > upper_bound)).any(axis=1)]
    return df_rel_filtered
```

```python
df_rel_cleaned=detect_outliers(df_rel,num_cols)
print(df_rel_cleaned)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Plot box plot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 5, i)
    sns.boxplot(y=df_rel_cleaned[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```
